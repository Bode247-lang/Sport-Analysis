```python
import requests
import pandas as pd

# Replace with your actual API key
api_key = '017695e48cd84deeaa97ed59c58532e0'

# URL to get the match information for the World Cup 2018
url = 'http://api.football-data.org/v4/competitions/PL/matches'#'https://api.football-data.org/v4/competitions/2018/matches'

# Request headers with API key
headers = {
    'X-Auth-Token': api_key
}

def fetch_world_cup_matches():
    try:
        # Send a GET request to the matches endpoint
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Raise HTTPError for bad responses (4xx and 5xx)

        # Parse the JSON response
        matches_data = response.json()

        # Print fetched data for debugging
        print('Fetched data:', matches_data)

        # Return the matches data
        return matches_data

    except requests.RequestException as e:
        print(f'Error fetching World Cup matches: {e}')
        return None

def save_matches_to_csv(matches_data, filename='world_cup_2018_matches.csv'):
    if matches_data and 'matches' in matches_data:
         # Create a list of dictionaries for DataFrame conversion
        matches_list = []
        for match in matches_data['matches']:
            match_info = {
                'Match ID': match.get('id'),
                'Home Team': match.get('homeTeam', {}).get('name'),
                'Away Team': match.get('awayTeam', {}).get('name'),
                'Home Score': match.get('score', {}).get('fullTime', {}).get('home'),
                'Away Score': match.get('score', {}).get('fullTime', {}).get('away'),
                'Match Date': match.get('utcDate')
            }
            matches_list.append(match_info)

        # Check if matches_list is empty
        if not matches_list:
            print('No matches found in the data.')
            return

        # Convert to DataFrame
        df = pd.DataFrame(matches_list)

        # Print DataFrame for debugging
        print('DataFrame preview:\n', df.head())
               # Save DataFrame to CSV
        df.to_csv(filename, index=False)
        print(f'Successfully saved matches data to {filename}')
    else:
        print('No match data to save.')

# Fetch the World Cup 2018 matches
matches_data = fetch_world_cup_matches()

# Save the match data to a CSV file
save_matches_to_csv(matches_data)
```
